```python
import pandas as pd
import matplotlib.pyplot as plt

### Loading the dataset from a csv file

df = pd.read_csv('iris_dataset.csv')
print(df.to_string())
```

```python
#Print the dataset with the *sepal length* values sorted in descending order
df_sorted_sepal_length = df.sort_values(by = 'sepal length (cm)', ascending = False)
print(df_sorted_sepal_length.to_string())
```

* Create line plots for each feature on a single plot, with proper labels and a legend at top-right corner. Label the X-axis as 'Index' and the Y-axis as 'Length (cm)'

```python
sepal_length = df['sepal length (cm)'].to_numpy()
sepal_width = df['sepal width (cm)'].to_numpy()
petal_length = df['petal length (cm)'].to_numpy()
petal_width = df['petal width (cm)'].to_numpy()
target = df['target'].to_numpy()

#print("sepal length:\n", sepal_length)
#rint("sepal width:\n",sepal_width)
#print("petal length:\n", petal_length)
#print("petal width:\n", petal_width)
#print("target:\n", target)

plt.plot(sepal_length, label = 'sepal length', color = "blue")
plt.plot(sepal_width, label = 'sepal width', color = "black")
plt.plot(petal_length, label = 'petal length', color = "red")
plt.plot(petal_width, label = 'petal width', color = "green")
#plt.plot(target)
plt.legend(loc='upper right')
plt.xlabel("Index")
plt.ylabel("Length (cm)")
plt.figure(figsize=(100, 100))
plt.show()
```

Now create a scatter plot between *petal length* and *petal width* with the following specifications:
- For points with target = 0, label should be 'Setosa' and color should be 'red'
- For points with target = 1, label should be 'Versicolor' and color should be 'blue'
- For points with target = 2, label should be 'Virginica' and color should be 'green'
- X axis should be labelled 'Petal Length'
- Y axis should be labelled 'Petal Width'
- Legend should be at top right corner

```python
# Plot for target = 0 (Setosa) - Red color
plt.scatter(petal_length[target == 0], petal_width[target == 0], label='Setosa', color='red')

# Plot for target = 1 (Versicolor) - Blue color
plt.scatter(petal_length[target == 1], petal_width[target == 1], label='Versicolor', color='blue')

# Plot for target = 2 (Virginica) - Green color
plt.scatter(petal_length[target == 2], petal_width[target == 2], label='Virginica', color='green')

# Set axis labels
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

# Add title
plt.title('Scatter Plot of Petal Length vs Petal Width')

# Add a legend at the top right corner
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
